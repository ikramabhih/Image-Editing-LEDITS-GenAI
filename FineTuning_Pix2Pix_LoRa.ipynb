```python
import torch
from diffusers import StableDiffusionInstructPix2PixPipeline, DDPMScheduler
from peft import LoraConfig, get_peft_model
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import pandas as pd
import os
from tqdm import tqdm
import gc
```

```python
import sys
print("Python utilis√© :", sys.executable)

!{sys.executable} -m pip install peft==0.13.0
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"üñ•Ô∏è  Device: {device}")

if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

# Chemins
DATASET_ROOT = r"C:\Users\user\dataset_leditspp_256"
TRAIN_DIR = os.path.join(DATASET_ROOT, "train")
VAL_DIR = os.path.join(DATASET_ROOT, "val")
TEST_DIR = os.path.join(DATASET_ROOT, "test")

# Param√®tres optimis√©s
IMAGE_SIZE = 256
BATCH_SIZE = 1
GRADIENT_ACCUMULATION = 4
NUM_EPOCHS = 10
LEARNING_RATE = 1e-4
DATASET_FRACTION = 0.15  # 15% du dataset
```

```python
# 2. CHARGEMENT DATASET (15%)
# ========================================
print("\nüìÇ Chargement du dataset...")

# Fonction pour charger uniquement metadata_fixed.tsv
def load_metadata(split_dir):
    tsv_path = os.path.join(split_dir, "metadata_fixed.tsv")

    if os.path.exists(tsv_path):
        print(f"   Loading: {tsv_path}")
        return pd.read_csv(tsv_path, sep="\t")
    else:
        raise FileNotFoundError(f"metadata_fixed.tsv introuvable dans {split_dir}")

# Charger les m√©tadonn√©es
train_meta = load_metadata(TRAIN_DIR)
val_meta = load_metadata(VAL_DIR)

print(f"   Dataset complet - Train: {len(train_meta)} | Val: {len(val_meta)}")

# S√©lectionner 15% (√©chantillonnage al√©atoire)
train_meta = train_meta.sample(frac=DATASET_FRACTION, random_state=42).reset_index(drop=True)
val_meta = val_meta.sample(frac=DATASET_FRACTION, random_state=42).reset_index(drop=True)

print(f"   15% s√©lectionn√© - Train: {len(train_meta)} | Val: {len(val_meta)}")
```

```python
# 3. DATASET CLASS
# ========================================
class EditDataset(Dataset):
    def __init__(self, root_dir, df, size=256):
        self.root = root_dir
        self.df = df
        self.transform = transforms.Compose([
            transforms.Resize((size, size), interpolation=transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize([0.5]*3, [0.5]*3)
        ])
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        
        # D√©tection automatique des colonnes
        if 'original_image_path' in row:
            orig_path = row['original_image_path']
            edit_path = row['edited_image_path']
            prompt = row['edit_prompt']
        else:
            # Adapter selon votre structure r√©elle
            orig_path = os.path.join("original", f"{row['id']}.png")
            edit_path = os.path.join("edited", f"{row['id']}.png")
            prompt = row['prompt']
        
        # Charger images
        orig_full = os.path.join(self.root, orig_path)
        edit_full = os.path.join(self.root, edit_path)
        
        try:
            orig = Image.open(orig_full).convert("RGB")
            edit = Image.open(edit_full).convert("RGB")
        except Exception as e:
            print(f"  Erreur chargement: {orig_full}")
            # Retourner image noire en fallback
            orig = Image.new("RGB", (IMAGE_SIZE, IMAGE_SIZE), (0, 0, 0))
            edit = Image.new("RGB", (IMAGE_SIZE, IMAGE_SIZE), (0, 0, 0))
        
        return {
            "original": self.transform(orig),
            "edited": self.transform(edit),
            "prompt": str(prompt)
        }

# Cr√©er datasets
train_dataset = EditDataset(TRAIN_DIR, train_meta, IMAGE_SIZE)
val_dataset = EditDataset(VAL_DIR, val_meta, IMAGE_SIZE)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, num_workers=0)

print(f"‚úÖ DataLoaders pr√™ts")
```

```python
# 4. CHARGEMENT MOD√àLE
# ========================================
print("\nü§ñ Chargement du mod√®le InstructPix2Pix...")
print("   (Premi√®re fois = t√©l√©chargement ~5.5 GB, patientez...)")

pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
    "timbrooks/instruct-pix2pix",
    torch_dtype=torch.float16,
    safety_checker=None,
    revision="fp16"
)

unet = pipe.unet
vae = pipe.vae.to(device)
text_encoder = pipe.text_encoder.to(device)
tokenizer = pipe.tokenizer
scheduler = pipe.scheduler

# Geler VAE et Text Encoder
vae.eval()
vae.requires_grad_(False)
text_encoder.eval()
text_encoder.requires_grad_(False)

print("‚úÖ Mod√®le charg√©")
```

```python
# 5. CONFIGURATION LORA
# ========================================
print("\nüîß Configuration LoRA...")

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["to_q", "to_k", "to_v", "to_out.0"],
    lora_dropout=0.1,
    bias="none"
)

unet = get_peft_model(unet, lora_config)
unet.to(device)
unet.enable_gradient_checkpointing()

trainable_params = sum(p.numel() for p in unet.parameters() if p.requires_grad)
total_params = sum(p.numel() for p in unet.parameters())
print(f"   Param√®tres entra√Ænables: {trainable_params/1e6:.2f}M / {total_params/1e6:.2f}M ({100*trainable_params/total_params:.2f}%)")

optimizer = torch.optim.AdamW(unet.parameters(), lr=LEARNING_RATE)
scaler = torch.cuda.amp.GradScaler()
```

```python

# ========================================
# 6. FONCTION DE LOSS
# ========================================
def compute_loss(batch):
    with torch.cuda.amp.autocast():
        edited = batch["edited"].to(device, dtype=torch.float16)
        original = batch["original"].to(device, dtype=torch.float16)
        
        with torch.no_grad():
            # Encode images ‚Üí latents
            latents = vae.encode(edited).latent_dist.sample() * vae.config.scaling_factor
            source_latents = vae.encode(original).latent_dist.sample() * vae.config.scaling_factor
            
            # Ajouter bruit
            noise = torch.randn_like(latents)
            bsz = latents.shape[0]
            timesteps = torch.randint(0, scheduler.config.num_train_timesteps, (bsz,), device=device).long()
            noisy_latents = scheduler.add_noise(latents, noise, timesteps)
            
            # Concat√©ner source + noisy (8 channels)
            model_input = torch.cat([source_latents, noisy_latents], dim=1)
            
            # Encode prompt
            text_inputs = tokenizer(
                batch["prompt"],
                padding="max_length",
                max_length=tokenizer.model_max_length,
                truncation=True,
                return_tensors="pt"
            )
            encoder_hidden_states = text_encoder(text_inputs.input_ids.to(device))[0]
        
        # Forward UNet (avec LoRA)
        noise_pred = unet(model_input, timesteps, encoder_hidden_states).sample
        
        # Loss MSE
        loss = torch.nn.functional.mse_loss(noise_pred, noise)
    
    return loss
```

```python
# 7. ENTRA√éNEMENT
# ========================================
print("\nüöÄ D√âBUT DE L'ENTRA√éNEMENT\n")

best_val_loss = float('inf')

for epoch in range(NUM_EPOCHS):
    print(f"{'='*60}")
    print(f"EPOCH {epoch+1}/{NUM_EPOCHS}")
    print(f"{'='*60}")
    
    # --- TRAIN ---
    unet.train()
    train_loss = 0
    optimizer.zero_grad()
    
    progress_bar = tqdm(train_loader, desc="Training")
    for step, batch in enumerate(progress_bar):
        loss = compute_loss(batch)
        loss = loss / GRADIENT_ACCUMULATION
        
        scaler.scale(loss).backward()
        
        if (step + 1) % GRADIENT_ACCUMULATION == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
        
        train_loss += loss.item() * GRADIENT_ACCUMULATION
        progress_bar.set_postfix({"loss": f"{loss.item()*GRADIENT_ACCUMULATION:.4f}"})
        
        # Nettoyage m√©moire
        if step % 50 == 0:
            torch.cuda.empty_cache()
    
    avg_train_loss = train_loss / len(train_loader)
    print(f"üìä Train Loss: {avg_train_loss:.4f}")
    
    # --- VALIDATION ---
    unet.eval()
    val_loss = 0
    
    with torch.no_grad():
        for batch in tqdm(val_loader, desc="Validation"):
            loss = compute_loss(batch)
            val_loss += loss.item()
    
    avg_val_loss = val_loss / len(val_loader)
    print(f"üìä Val Loss: {avg_val_loss:.4f}")
    
    # --- SAUVEGARDE ---
    save_dir = f"lora_pix2pix_epoch{epoch+1}"
    os.makedirs(save_dir, exist_ok=True)
    unet.save_pretrained(save_dir)
    print(f"üíæ Sauvegard√©: {save_dir}")
    
    # Sauvegarder le meilleur mod√®le
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        best_dir = "lora_pix2pix_best"
        os.makedirs(best_dir, exist_ok=True)
        unet.save_pretrained(best_dir)
        print(f"‚≠ê MEILLEUR MOD√àLE sauvegard√©: {best_dir}")
    
    print()
    
    # Nettoyage
    torch.cuda.empty_cache()
    gc.collect()

print("="*60)
print("‚úÖ ENTRA√éNEMENT TERMIN√â!")
print(f"üìÅ Mod√®les sauvegard√©s dans:")
print(f"   - lora_pix2pix_epoch1/")
print(f"   - lora_pix2pix_epoch2/")
print(f"   - lora_pix2pix_epoch3/")
print(f"   - lora_pix2pix_best/ (meilleur mod√®le)")
print("="*60)
```
